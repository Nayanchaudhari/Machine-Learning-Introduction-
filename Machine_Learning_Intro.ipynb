Question 1: Explain the differences between AI, ML, Deep Learning (DL), and Data Science (DS).
 - 1. Artificial Intelligence (AI)
   - Definition: AI is the broadest field. It’s about creating machines or systems that can perform tasks that normally require human intelligence (reasoning, problem-solving, learning, decision-making).

   - Examples:- Self-driving cars, Recommendation systems (Netflix, Amazon).

 - 2. Machine Learning (ML)
   - Definition: A subset of AI. It enables machines to learn from data and improve their performance over time without being explicitly programmed.

   - Examples:- Spam email detection, Predicting stock prices.

 - 3. Deep Learning (DL)
   - Definition: A subset of ML that uses artificial neural networks with many layers (hence "deep").

   - Examples:- Face recognition (Facebook, iPhone), Voice assistants (Alexa, Siri).

- 4. Data Science (DS)
   - Definition: An interdisciplinary field that focuses on extracting meaningful insights from data. It combines statistics, mathematics, programming, ML/AI, and domain knowledge.

   - Examples:- Business dashboards (Power BI, Tableau), Customer churn analysis.     


Question 2: What are the types of machine learning? Describe each with one real-world example.
 - Machine Learning (ML) is broadly divided into four main types.

 - 1. Supervised Learning
   - Definition: The model is trained on a labeled dataset (input data with the correct output already known).

   - Goal: Learn a mapping from inputs → outputs and make predictions on new data.

   - Techniques: Regression, Classification.

   - Example:- Spam Email Detection: Input = email text, Output = "spam" or "not spam". The model learns from labeled examples to classify new emails.

 - 2. Unsupervised Learning
   - Definition: The model is given unlabeled data (no predefined outputs).

   - Goal: Find hidden patterns, structure, or groupings in data.

   - Techniques: Clustering, Dimensionality Reduction.

   - Example:- Customer Segmentation: E-commerce sites group customers based on behavior (purchase history, browsing patterns) to create targeted marketing strategies.

 - 3. Semi-Supervised Learning

   - Definition: The model is trained on a small amount of labeled data and a large amount of unlabeled data.

   - Goal: Use unlabeled data to improve learning accuracy when labeling is expensive or time-consuming.

   - Example:- Medical Diagnosis: Doctors label only a few X-ray scans (normal/abnormal), while thousands of unlabeled scans help the model learn patterns to detect diseases more accurately.

 - 4. Reinforcement Learning (RL)

   - Definition: The model (agent) learns by interacting with an environment, receiving rewards or penalties for its actions.

   - Goal: Learn the best strategy (policy) to maximize cumulative rewards.

   - Techniques: Q-learning, Deep Q-Networks.

   - Example:- Self-Driving Cars: The car (agent) gets positive rewards for staying in the lane and avoiding obstacles, and penalties for crashes or breaking traffic rules.


Question 3: Define overfitting, underfitting, and the bias-variance tradeoff in machine learning.

 - 1. Overfitting
   - Definition: When a model learns too much detail and noise from the training data, performing very well on training but poorly on new/unseen data.

   - Example:- A student memorizes answers word-for-word from past exams → scores high in practice tests but fails in real exams with slightly different questions.

   - Result: High variance, low bias

 - 2. Underfitting
   - Definition: When a model is too simple to capture the underlying patterns in the data, leading to poor performance on both training and test data.

   - Example:- A student skims notes without real understanding → performs badly on both practice and real exams.  

   - Result: High bias, low variance.

 - 3. Bias-Variance Tradeoff
   - Bias: Error due to simplifying assumptions in the model.

     - High bias → underfitting.

   - Variance: Error due to model being too sensitive to small fluctuations in training data.

     - High variance → overfitting.  
  
   - Tradeoff:
     - Increasing model complexity reduces bias but increases variance.
     - Decreasing complexity reduces variance but increases bias.
     - Goal = Find the sweet spot where both bias and variance are balanced → good generalization.


Question 4: What are outliers in a dataset, and list three common techniques for handling them.

 - Definition :- Outliers are data points that lie far away from the general distribution of a dataset.

 - Three common techniques for handling them.
   - 1. Removal (Deleting Outliers)
      - When to use:- If outliers are due to clear data entry or measurement errors, remove them.

      - Example: Age = 500 in a human dataset → invalid, so remove.
   - 2. Transformation
      - When to use:- Apply mathematical transformations to reduce the impact of outliers.

      - Example: Incomes ranging from ₹20,000 to ₹10,00,000 → log transformation reduces skewness.

   - 3. Capping
      - When to use:- If outliers are extreme but shouldn’t be removed entirely.

      - Example: Salary above 99th percentile capped at the 99th percentile value.   


Question 5: Explain the process of handling missing values and mention one imputation technique for numerical and one for categorical data.

 - Process of Handling Missing Values.
   - 1. Identify missing values
      - Check using functions like .isnull() or .info() in Pandas.
   - 2. Analyze the pattern
      - Missing completely at random (MCAR)
      - Missing at random (MAR)
      - Missing not at random (MNAR)

   - 3. Decide on a strategy
     - Remove: Drop rows/columns if missing values are too many.
     - Impute: Replace missing values with estimated values.

 - Imputation Techniques
   - For Numerical Data → Mean/Median Imputation
     - Replace missing numeric values with the mean (if data is normally distributed) or median (if data is skewed).

     - Example: If a dataset of student ages has missing values → fill them with the median age.

   - For Categorical Data → Mode Imputation
     - Replace missing categorical values with the most frequent category (mode).

     - Example: If “Gender” has missing values, fill them with the most common value (e.g., “Male” or “Female”).

```python
#Question 8: Write a Python script to:
#Generate 1000 samples from a normal distribution.
#Introduce 50 random missing values.
#Fill missing values with the column mean.
#Plot a histogram before and after imputation.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Generate 1000 samples from a normal distribution.
np.random.seed(42)
data = np.random.normal(1,5,1000)

#Introduce 50 random missing values.
missing_value = data.copy()
missing_indices = np.random.choice(1000,50,replace=False)
missing_value[missing_indices] = np.nan

# Convert to DataFrame
df = pd.DataFrame(missing_value,columns=["Values"])
df.isnull().sum()

#Fill missing values with the column mean.
mean_value = df["Values"].mean()
df["Imputed"] = df["Values"].fillna(mean_value)

#Plot a histogram before and after imputation.
plt.figure(figsize=(12,5))

# Histogram before imputation
plt.subplot(1,2,1)
plt.hist(df["Values"].dropna() ,bins=30, color="skyblue")
plt.title("Before Imputation")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Histogram after imputation
plt.subplot(1,2,2)
plt.hist(df["Imputed"],bins=30,color="lightgreen")
plt.title("After Imputation (Mean Filled)")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```

```python
#Question 10: You are working as a data scientist for a retail company. You receive a customer transaction dataset that contains:

#● Missing ages, ● Outliers in transaction amount, ● A highly imbalanced target (fraud vs. non-fraud), ● Categorical variables like payment method.

#Explain the step-by-step data preparation plan you’d follow before training a machine learning model. Include how you’d address missing data, outliers, imbalance, and encoding.



import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, RobustScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, average_precision_score


# 1. Simulate retail dataset

# Fraud detection dataset (imbalanced)
X, y = make_classification(n_samples=5000, n_features=5,
                           n_informative=3, n_redundant=0,
                           weights=[0.95, 0.05],  # 5% fraud
                           random_state=42)

df = pd.DataFrame(X, columns=["age", "transaction_amount", "feature1", "feature2", "feature3"])
df["fraud"] = y

# Add categorical column (payment method)
np.random.seed(42)
df["payment_method"] = np.random.choice(["CreditCard", "DebitCard", "UPI", "Wallet"], size=len(df))

# Introduce missing ages (about 10%)
missing_indices = np.random.choice(df.index, size=500, replace=False)
df.loc[missing_indices, "age"] = np.nan

# Add extreme outliers in transaction_amount
outlier_indices = np.random.choice(df.index, size=10, replace=False)
df.loc[outlier_indices, "transaction_amount"] = df["transaction_amount"] * 1000

print("Sample data:")
print(df.head(), "\n")
print("Class distribution:\n", df["fraud"].value_counts(normalize=True))


# 2. Feature Engineering

# Log-transform transaction_amount
df["transaction_amount_log"] = np.log1p(df["transaction_amount"])

# Add outlier flag
df["amount_outlier_flag"] = (df["transaction_amount"] > df["transaction_amount"].quantile(0.999)).astype(int)


# 3. Train/Test Split

X = df.drop(columns=["fraud"])
y = df["fraud"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)


# 4. Preprocessing Pipelines

numeric_features = ["age", "transaction_amount_log", "feature1", "feature2", "feature3", "amount_outlier_flag"]
categorical_features = ["payment_method"]

num_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", RobustScaler())
])

cat_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="constant", fill_value="missing")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", num_pipeline, numeric_features),
        ("cat", cat_pipeline, categorical_features)
    ]
)


# 5. Model Pipeline

clf = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("classifier", RandomForestClassifier(
        class_weight="balanced",  # handles imbalance
        n_estimators=200,
        random_state=42
    ))
])


# 6. Training & Evaluation

clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
y_proba = clf.predict_proba(X_test)[:,1]

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("Average Precision Score (PR-AUC):", average_precision_score(y_test, y_proba))
```
